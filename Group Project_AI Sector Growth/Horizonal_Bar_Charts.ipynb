<h2>Data Cleaning</h2>

```python
df.isna().sum()
```

This tells us there are null values in various rows. But the only important cols that cannot have null are 'AI_Growth' and 'Patents_Growth'

```python
df = df.dropna(subset=['AI_Growth','Patents_Growth']) #we drop the rows that have null values ONLY in those 2 rows
```

```python
#filtering out the important data
selected_sectors = ['Software & IT Services', 'Industrial Goods', 'Banking & Investment Services', 'Healthcare Services & Equipment', 'Pharmaceuticals & Medical Research']
number_metrics = ['AI_Publications', 'AI patents']
growth_metrics = ['AI_Growth', 'Patents_Growth']
analysis_metrics = ['AI_Growth', 'Patents_Growth', 'AI_Publications', 'AI patents']
```

<h1>Horizontal Bar Graphs (Used in Slides)</h1>

```python
sector_colors = {
    'Software & IT Services': '#A6D8E7',
    'Industrial Goods': '#5FB8B2', 
    'Banking & Investment Services': '#3A7CA5',
    'Healthcare Services & Equipment': '#7E6CAB' , 
    'Pharmaceuticals & Medical Research': '#4F6272', 
}
```

Patents Growth - Patent + Growth Metric:

```python
pGrowth_analysis = df.groupby('Sector').Patents_Growth.mean().round(2).loc[selected_sectors].sort_values(ascending=True)
colors = [sector_colors[sector] for sector in pGrowth_analysis.index]
pGrowth_analysis.plot(
    kind='barh',
    title='Average AI Patent Growth across Sectors',
    xlabel='Average Patent Growth',
    ylabel='-----Sector Rank----->',
    color=colors,
    figsize=(6,4)
)
```

AI Publication - Publication + Number metric:

```python
AIpub_analysis = df.groupby('Sector').AI_Publications.mean().round(2).loc[selected_sectors].sort_values(ascending=True)
colors = [sector_colors[sector] for sector in AIpub_analysis.index]
AIpub_analysis.plot(
    kind='barh',
    title='Average Number of AI Publication across Sectors',
    xlabel='Average AI Publication',
    ylabel='-----Sector Rank----->',
    color=colors,
    figsize=(6,4)
)
```

Publication growth - Publication + Growth metric:

```python
aiG_analysis = df.groupby('Sector').AI_Growth.mean().round(2).loc[selected_sectors].sort_values(ascending=True)
colors = [sector_colors[sector] for sector in aiG_analysis.index]
aiG_analysis.plot(
    kind='barh',
    title='Average AI Publications Growth across Sectors',
    xlabel='Average AI Growth',
    ylabel='-----Sector Rank----->',
    color=colors,
    figsize=(6,4)
)
```

AI Patent - Patent + Number metric:

```python
aiPat_analysis = df.groupby('Sector')['AI patents'].mean().round(2).loc[selected_sectors].sort_values(ascending=True)
colors = [sector_colors[sector] for sector in aiPat_analysis.index]
aiPat_analysis.plot(
    kind='barh',
    title='Average Number of AI Patents across Sectors',
    xlabel='Average AI Patents',
    ylabel='-----Sector Rank----->',
    color=colors,
    figsize=(6,4)
)
```
